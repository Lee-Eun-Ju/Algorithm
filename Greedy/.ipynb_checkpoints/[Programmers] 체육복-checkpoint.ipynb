### 체육복

#### 1. 문제 이해

- 제한사항
1) 전체 학생 수는 2명 이상 30명 이하
2) 도난당한 학생수는 2명 이상 n명 이하
3) 여벌의 체육복이 있는 학생수는 1명 이상 n명 이하

- 인간과 달리 컴퓨터가 문제 해결을 하도록 하기 위해서는  
  단계별로 해결 방법을 고려해보는 습관을 가져야 한다.

#### 2. 탐욕법(Greedy)

- 알고리즘의 각 단계에서 그 순간에 최적이라고 생각되는 것을 선택
- 탐욕법으로 최적해를 찾을 수 있는 문제 : 현재의 선택이 마지막 해답의 최적성을 해치지 않는 경우

- 탐욕법 적용 가능성 확인

  

- 빌려줄 학생들을 정해진 순서로 살펴야 하고, 정해진 순서에 따라 우선하여 빌려줄 방향을 정해야 함

##### 방법(1)
- (착안점) 학생의 수는 기껏해야 30!
- 1 학생 수만큼 배열을 확보하고, 여기에 각자가 가지고 있는 체육복 수 기록
- 2 번호 순서대로 스캔하면서 빌려줄 관계 지정
- (복잡도) reserve 길이 + lost 길이 + 전체 학생수 비례 => O(n)

##### 방법(2)
- (착안점) 만약 전체 학생 수가 매우 크다면? 그런데 여벌의 체육복을 가져온 학생은 매우 적다면?
- 1 여벌의 체육복을 가져온 학생들의 번호(reserve)를 정렬하고,   
- 2 하나 하나 순서대로 살펴보면서 빌려줄 수 있는 학생을 찾아 처리 => 해시적용(상수시간 처리)
- (복잡도) reserve 길이가 k일 때 O(klogk) + 해시 적용으로 O(k)*O(1)

#### 3. 문제 풀이

- 강의 수강 전 문제풀이

```python
n = 10
lost = [1,5,8,3,7]
reserve = [7,5,2,9,6]

# 강의 수강 전 문제풀이
def solution(n, lost, reserve):

    # n:학생수, lost:도난당한 학생, reserve:여벌옷이 있는 학생(2개)
    # 여벌 옷 주기 전 기록
    data = [0]*(n+1)
    for i in range(1,n+1):
        if i in reserve and i in lost:
            data[i] = 1
        elif i in reserve:
            data[i] = 2
        elif i not in lost:
            data[i] = 1

    # 2벌 있는 친구가 0벌 있는 친구에게 주기
    for i in range(1,n+1):       
        if data[i]==0:
            if data[i-1]==2:
                data[i]+=1; data[i-1]-=1
            elif data[i+1]==2:
                data[i]+=1; data[i+1]-=1
    
    answer = data.count(1) + data.count(2)
    return answer

solution(n, lost, reserve)
```

- 강의 수강 후 Retry

```python
n = 10
lost = [1,5,8,3,7]
reserve = [7,5,2,9,6]

# 방법 2
def solution(n, lost, reserve):

    reserve.sort()

    dict = {}
    for i in lost:
        if i in reserve:
            reserve.remove(i)
            continue
        dict[i] = 0
           
    for i in reserve:
        if (i-1) in dict:
            dict.pop(i-1)
        elif (i+1) in dict:
            dict.pop(i+1)
    
    answer = n - len(dict)
    return answer

solution(n, lost, reserve)
```

- 강사의 문제 풀이

```python
n = 10
lost = [1,5,8,3,7]
reserve = [7,5,2,9,6]

# 방법1
def solution(n, lost, reserve):
    # 1번 앞, n번 뒤 -> 학생이 있는 것처럼 만들어
    u = [1]*(n+2) 
    for i in reserve:
        u[i] += 1
    for i in lost:
        u[i] -= 1

    # 빌려주는 규칙 적용 -> 단순하게 코딩 가능
    for i in range(1,n+1):
        if u[i-1] ==0 and u[i]==2:
            u[i-1:i+1] = [1,1]
        elif u[i] ==2 and u[i+1]==0:
            u[i:i+2] = [1,1] 

    return len([x for x in u[1:-1] if x>0])

solution(n, lost, reserve)
```

```python
n = 10
lost = [1,5,8,3,7]
reserve = [7,5,2,9,6]

# 방법2 (해시 - set 이용) 
# set 자료구조 : hash table을 통해 구현하므로 상수 복잡도 접근
# <-> dictionary와의 차이 : set은 key만 존재, value는 없다

def solution(n, lost, reserve):
    
    # & : 교집합을 저장, - : 차집합을 저장
    s = set(lost) & set(reserve) # 2개 있으나 1개 도난당한 학생
    l = set(lost) - s # 체육복을 빌려야 하는 학생
    r = set(reserve) - s # 체육복을 빌려줄 수 있는 학생 

    for x in sorted(r):
        if x-1 in l:
            l.remove(x-1)
        elif x+1 in l:
            l.remove(x+1)
    
    return n - len(l) 

solution(n, lost, reserve)
```
