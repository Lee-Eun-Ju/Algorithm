### 가장 큰수

#### 1. 문제 이해
- 제한사항
- 1 numbers의 길이는 1 이상 100,000 이하
- 2 numbers의 원소는 0 이상 1,000 이하 => 4글자
- 3 문자열로 바꾸어 return


#### 2. 문제 해결 방법

##### 방법 1
- 1 빈 문자열로 수를 초기화
- 2 가장 크게 만들 수 있는 수를 선택
- 3 그 수를 현재 수에 이어 붙인다
- 4 모든 수를 다 사용할 때까지 반복

- (복잡도) 수 선택(목록의 길이에 비례) * 반복(목록의 길이에 비례) => O(n^2)

##### 방법 2
- 1 빈 문자열로 수를 초기호
- 2 수의 목록을 정렬 (크게 만드는 것 우선으로) => 정렬에 사용되는 기준!
- 3 목록에서 하나씩 꺼내어 현재 수에 이어 붙인다
- 4 모든 수를 다 사용할 때까지 반복

##### 크게 만드는 수의 기준  

* 3 vs 32 => 332으로 3 먼저!
* 3 vs 33 => 아무거나
* 3 vs 34 => 343으로 34 먼저!

- 34 vs 342 => 34234 34342 => 34 먼저!
- 34 vs 343 => 34343 34334 => 34 먼저 => 제한사항 4글자까지 비교
- 34 vs 344 => 34344 34434 => 344 먼저!



#### 3. 문제 풀이

- 강의 수강 전 문제 풀이

```python
numbers = [6,10,2]
numbers = [3,30,34,5,9]


def solution(numbers):
    import itertools 
    numbers = list(map(str, numbers))
    comb_num = list(itertools.permutations(numbers, len(numbers)))
    result = list(map(int, ["".join(i) for i in comb_num]))
    answer = str(max(result))
    return answer


def solution(numbers):
    numbers = list(map(str, numbers))
    
    word = []*len(numbers)
    for i in numbers:
        if len(i)==1:
            a=int(i[0]); b=-1; c=-1
        elif len(i)==2:
            a=int(i[0]); b=int(i[1]); c=-1
        elif len(i)==3:
            a=int(i[0]); b=int(i[1]); c=int(i[2])
        word.append((i, a, b, c))   
    word = sorted(word, key=lambda x : (x[1], x[2], x[3]), reverse=True)

    answer = ""
    for j in word:
        answer += j[0]

    return answer
```

- 강의 수강 후 Retry

```python
numbers = [3,30,34,5,9]

def solution(numbers):
    num = list(map(str, numbers))
    num.sort(key= lambda x: (x*4)[:4], reverse=True)

    answer = "".join(num)
    return answer

solution(numbers)
```

- 강사의 풀이

```python
numbers = [3,30,34,5,9]

def solution(numbers):
    numbers = [str(x) for x in numbers]
    numbers.sort(key=lambda x: (x*4)[:4], reverse=True)

    # 제한사항 : 0이상 1000이하 
    # 0만 2개 이상 있는 배열인 경우 => numbers =[0, 0, 0], answer="000"이 아닌 "0"
    if numbers[0]=="0":
        answer = "0" 
    else:
        answer = "".join(numbers)
    return answer

solution(numbers)
```

- (복잡도) 문자열변경(n) + 정렬(nlogn) + join(n) => O(nlogn)
