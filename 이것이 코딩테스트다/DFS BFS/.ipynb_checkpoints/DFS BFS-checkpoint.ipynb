## DFS/BFS

### 1. 꼭 필요한 자료구조 기초

- 탐색 : 많은 양의 데이터 중 원하는 데이터를 찾는 과정
- 탐색 알고리즘 : DFS, BFS 

- 자료 구조 : 데이터를 표현하고 관리하고 처리하기 위한 구조
  - 삽입(Push) : 데이터를 삽입하다.
  - 삭제(Pop) : 데이터를 삭제하다.
        
  - 오버플로(Overflow) : 특정한 자료구조가 수용할 수 있는 데이터의 크기를 이미 가득 채운 상태에서, 삽입 연산하는 경우
  - 언더플로(Underflow) : 특정한 자료구조에 데이터가 전혀 들어 있지 않은 상태에서, 삭제 연산을 수행하는 경우

- 스택 : 박스 쌓기 (선입후출 구조 또는 후입선출 구조)
  - append : 리스트의 가장 뒤쪽에 데이터를 삽입
  - pop : 리스트의 가장 뒤쪽에서 데이터를 꺼냄

```python
stack = []

stack.append(5)
stack.append(2)
stack.append(3)
stack.append(7)
stack.pop()
stack.append(1)
stack.append(4)
stack.pop()

print(stack) #최하단 원소부터
print(stack[::-1]) #최상단 원소부터
```

- 큐 : 대기 줄 (공전한 자료구조, 선입선출 구조)
  - collections 모듈의 deque 자료구조 활용
  - append(), popleft(), popright() : 스택과 큐의 장점을 모두 채택

```python
from collections import deque

# 큐 구현을 위해서, deque 이요어
queue = deque() 

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
queue.popleft()
queue.append(1)
queue.append(4)
queue.popleft()

print(queue) #순서대로
queue.reverse(); print(queue) #역순대로

print(list(queue)) #list 자료형
```

- 재귀 함수 : 자기 자신을 다시 호출하는 함수
  - 스택 자료구조 이용 (선입후출) : 가장 마지막에 호출한 함수가 수행을 끝내야 그 앞의 함수 호출이 종료
  - 프랙털 구조와 흡사하며, 실제로 프랙털 이미지 출력하는 프로그램 작성시 이용

```python
def recursive_function():
    print("재귀 함수를 호출합니다.")  
    # 함수에서 자기 자신을 다시 호출
    recursive_function() 

# 문자열 무한히 출력 - 자기 자신을 계속해서 추가로 불러오기 때문
recursive_function() 
```

  - 재귀 함수의 종료 조건
    - 문제 풀이시, 재귀 함수가 언제 끝날지 종료 조건을 꼭 명시!
    - if문이 종료 조건 역할 수행

```python
def recursive_function(i):
    # 100번째 출력했을 때 종료
    if i==100:
        return
    print(i,'번째 재귀 함수에서',i+1,'번째 재귀 함수를 선출합니다.')
    recursive_function(i+1)
    print(i,'번째 재귀 함수를 종료합니다.')

recursive_function(1)
```

- ex) 팩토리얼 문제 : 반복 구현 vs 재귀 구현
  - 재귀 구현 : 수학의 점화식을 소스코드로 그대로 옮길 수 있어 간결
    
  - 수학의 점화식 : 특정 함수를 자신보다 더 작은 변수에 대한 함수와의 관계로 표현한 것
    - n이 0 혹은 1일 때 : factorial(n)=1
    - n이 1보다 클 때 : factorial(n)=n * factorial(n-1) 
    - 이때, n에 대한 조건 : 재귀함수의 종료 조건

```python
def factorial_iterative(n):
    result =1
    for i in range(1,n+1):
        result *= i
    return result

def factorial_recursive(n):
    if n<=1:
        return 1  
    return n * factorial_recursive(n-1)
    
print(factorial_iterative(5))
print(factorial_recursive(5))
```

## 2. 탐색 알고리즘 DFS/BFS

### 1) DFS (Depth-First Search, 깊이 우선 탐색)
- 그래프에서 깊은 부분을 우선적으로 탐색하는 알고리즘
- 스택 자료구조 (선입후출), 재귀함수 이용하여 구현

- 그래프의 기본 구조

  <img src=https://blog.kakaocdn.net/dn/4KhlW/btq1AGGdauX/ziMkE3TdIKSzvJ8HFUy27k/img.png width="250" align="left">
  
  - 그래프는 노드(=정점), 간선으로 표현된다.  
    (간선으로 연결되어 있다=두 노드는 인접하다)
    
    
  - 그래프 탐색이란 하나의 노드를 시작으로 다수의 노드를 방문하는 것  

- 프로그래밍 방식에서 그래프 표현 방법
    - 인접 행렬 : 2차원 배열로 그래프의 연결 관계를 표현
      - 연결이 되어 있지 않은 노드끼리는 무한의 비용이라고 작성
      - INF = 99999999
    - 인접 리스트 : 리스트로 그래프의 연결 관계를 표현
      - 모든 노드에 연결된 노드에 대한 정보를 차례대로 입력 
      - append() 이용
      
      
- 인접 행렬 vs 인접 리스트
  - 인접 행렬 : 모든 관계를 저장하므로, 노드 개수가 많을수록 메모리가 불필요하게 낭비
  - 인접 리스트 : 메모리를 효율적으로 사용하나, 특정 두 노드가 연결되어 있는지에 대한 정보를 얻는 속도가 느리다. 대신 특정 노드와 연결된 모든 인접 노드를 순회해야 하는 경우, 낭비가 적다

```python
# 인접 행렬 방식 
INF = 999999999
graph = [
    [0,7,5],  # 0이 0,1,2 노드에 연결된 간선
    [7,0,INF],
    [5,INF,0]
]
print(graph)

# 인접 리스트 방식 
graph = [[] for _ in range(3)]
graph[0].append((1,7)) # 연결된 노드, 간선
graph[0].append((2,5))
graph[1].append((0,7))
graph[2].append((0,5))
print(graph)
```

- DFS
  - 1) 탐색 시작 노드를 스택에 삽입하고 방문 처리(-> 각 노드를 한 번씩만 처리)
  - 2) 최상단 노드에 방문하지 않은 인접 노드가 있으면, 그 인접 노드를 스택에 넣고 방문 처리  
       - 방문하지 않은 인접 노드가 여러 개 있는 경우, 번호가 낮은 순서부터 처리  
       - 방문하지 않은 인접 노드가 없으면, 스택에서 최상단 노드를 꺼낸다.
       
  - 3) 2번의 과정을 더 이상 수행할 수 없을 때까지 반복

```python
def dfs(graph, v, visited):
    # 현재 노드를 방문처리
    visited[v]=True
    print(v, end=' ') # 노드의 탐색 순서 (스택에 들어간 순서) 출력
    
    # 그래프 구조의 노드들 탐색
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)

graph = [
    [], # 0 노드에 연결된 노드 (0노드 존재하지 X)
    [2,3,8], # 1 노드에 연결된 노드 (일반적으로 번호가 낮은 순서로)
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

visited = [False]*9  # 방문 정보 기록
dfs(graph, 1, visited)
```

### 2) BFS (Breath First Search, 너비 우선 탐색)
- 가까운 노드부터 탐색하는 알고리즘
- 큐 자료구조 (선입선출 방식)
- 실제 수행 시간은 일반적으로 DFS보다 좋은 편이다

- BFS
  - 1) 탐색 시장 노드를 큐에 삽입하고 방문 처리
  - 2) 큐에서 노드를 꺼내 해당 노드의 인접 노드 중에서 방문하지 않은 노드를 모두 큐에 삽입하고 방문 처리
  - 3) 2의 과정을 더 이상 수행할 수 없을 때까지 반복

```python
from collections import deque

def bfs(graph, start, visited):
    queue = deque([start]) # 큐 구현
    visited[start] =True # 현재 노드 방문 처리
    
    # 큐가 빌 때까지 반복
    while queue:
        v = queue.popleft()
        print(v, end=" ")
        
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i]=True

graph = [
    [], # 0 노드에 연결된 노드 (0노드 존재하지 X)
    [2,3,8], # 1 노드에 연결된 노드 (일반적으로 번호가 낮은 순서로)
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]
visited = [False]*9

bfs(graph, 1, visited)
```
